```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("orders (1).csv")
print(df.info())
```

```python
df.head()
```

```python
df["Qty."]= df["Qty."].str.split("/").str[0].astype(int)
```

```python
df = df[df["Product"]=="MIS"].reset_index(drop=True)
df
```

```python
## Different types of charges for Individual trade
```

```python
df["Turnover"]= df["Qty."] * df["Avg. price"].astype(float)
df
```

```python
df["Brokerage"] = np.where((df["Turnover"]*(0.0003))>20, 20 ,df["Turnover"]*0.0003).astype(int)
df
```

```python
df["STT/CTT"] = np.where((df["Type"]=="SELL"),df["Turnover"]*0.00025,0)
df["STT/CTT"] = df["STT/CTT"].round(2)
df
```

```python
df["ETC"]=df["Turnover"]*0.0000297
df["ETC"] = df["ETC"].round(2)
df
```

```python
df["SEBI"] = df["Turnover"]*(10/10000000)
df["SEBI"] = df["SEBI"].round(2)
df
```

```python
df["Stamp Charges"] = np.where((df["Type"]=="BUY"),df["Turnover"]*0.00003,0)
df["Stamp Charges"] = df["Stamp Charges"].round(2)
df
```

```python
df["GST"] = (df["Brokerage"] + df["SEBI"] + df["ETC"])*(18/100)
df["GST"] = df["GST"].round(2)
df
```

```python
df["Total Charges"] = df["Brokerage"] + df["STT/CTT"] + df["ETC"] + df["SEBI"] + df["Stamp Charges"] + df["GST"]
df["Total Charges"] = df["Total Charges"].round(2)
df
```

```python
df_sol1 = df
df_sol1
```

```python
## Stock wise and Type wise analysis with weighted Avg. price & calculated charges
```

```python
filtered = df[df["Instrument"].isin(["ASHOKLEY", "TATAMOTORS"]) & (df["Status"] == "COMPLETE")].copy()
filtered["Instrument_Type"] = filtered["Instrument"] + " " + filtered["Type"]
```

```python
df_sol2 = filtered.groupby(["Instrument", "Type"]).agg({
    "Qty.": "sum",
    "Avg. price": "mean",
    "Turnover": "sum",
    "Brokerage": "sum",
    "STT/CTT": "sum",
    "ETC": "sum",
    "SEBI": "sum",
    "GST": "sum",
    "Stamp Charges": "sum",
    "Total Charges": "sum"
}).reset_index()
df_sol2
```

```python
## Overall Summary of each Stocks
```

```python
## gross pnl =(sell price - buy price)* qty
```

```python
data = {
    'Instrument': ['ASHOKLEY', 'ASHOKLEY', 'TATAMOTORS', 'TATAMOTORS'],
    'Type': ['BUY', 'SELL', 'BUY', 'SELL'],
    'Qty': [5000, 5000, 250, 250],
    'Avg price': [125.662500, 125.953333, 490.550000, 492.100000]
}
df = pd.DataFrame(data)
df
```

```python
buy_df = df[df['Type'] == 'BUY'].set_index('Instrument')
sell_df = df[df['Type'] == 'SELL'].set_index('Instrument')
merged = buy_df[['Qty', 'Avg price']].rename(columns={'Qty': 'Buy_Qty', 'Avg price': 'Buy_Price'})\
.join(sell_df[['Qty', 'Avg price']].rename(columns={'Qty': 'Sell_Qty', 'Avg price': 'Sell_Price'}))
```

```python
merged['Gross_PnL'] = (merged['Sell_Price'] - merged['Buy_Price']) * merged['Sell_Qty']
merged['Gross_PnL'] = merged['Gross_PnL'].round(2)
merged
```

```python
## net pnl = Gross pnl - total charges
```

```python
data = {
    'Instrument': ['ASHOKLEY', 'ASHOKLEY', 'TATAMOTORS', 'TATAMOTORS'],
    'Type': ['BUY', 'SELL', 'BUY', 'SELL'],
    'Qty': [5000, 5000, 250, 250],
    'Avg price': [125.625, 125.953333, 490.55, 492.19],
    'Total Charges': [136.01, 251.06, 31.72, 58.81]
}
df = pd.DataFrame(data)
df
```

```python
buy = df[df['Type'] == 'BUY'].set_index('Instrument')
sell = df[df['Type'] == 'SELL'].set_index('Instrument')
merged = buy[['Qty', 'Avg price', 'Total Charges']].rename(columns={
    'Qty': 'Buy_Qty',
    'Avg price': 'Buy_Price',
    'Total Charges': 'Buy_Charges'
}).join(
    sell[['Qty', 'Avg price', 'Total Charges']].rename(columns={
        'Qty': 'Sell_Qty',
        'Avg price': 'Sell_Price',
        'Total Charges': 'Sell_Charges'
    })
)
```

```python
merged['Net_PnL'] = (merged['Sell_Price'] - merged['Buy_Price']) * merged['Sell_Qty'] - merged["Buy_Charges"]+ merged["Sell_Charges"]
merged['Net_PnL'] = merged['Net_PnL'].round(2)
merged
```

```python
# total Charge = buy charges + sell charges
```

```python
merged['Total_Charges'] = merged["Buy_Charges"]+ merged["Sell_Charges"]
merged['Total_Charges'] = merged['Total_Charges'].round(2)
merged
```

```python
# prc charge = (total charge/Gross pnl)*100
```

```python
merged['%Charges_on_Gross_PnL'] = (merged['Total_Charges'] / (merged['Sell_Price'] - merged['Buy_Price']) * merged['Sell_Qty'] ) * 100
merged['%Charges_on_Gross_PnL'] = merged['%Charges_on_Gross_PnL'].round(2)
merged
```

```python
## final table of ques. 3
```

```python
merged["Instrument"] = merged.index
merged
```

```python
merged["Gross_PnL"] = (merged['Sell_Price'] - merged['Buy_Price']) * merged['Sell_Qty']
merged
```

```python
df_sol3 = merged[["Instrument","Gross_PnL","Total_Charges","Net_PnL","%Charges_on_Gross_PnL"]]
df_sol3.head()
```
